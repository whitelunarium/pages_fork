  <h2>‚è± Runtime Complexity (Time)</h2>
  <p>Runtime complexity describes how the execution time of an algorithm <b>scales with input size <code>n</code></b>.</p>

  <p><b>Common complexities:</b></p>
  <ul>
    <li><b>O(1)</b> ‚Äî constant time</li>
    <li><b>O(log n)</b> ‚Äî logarithmic</li>
    <li><b>O(n)</b> ‚Äî linear</li>
    <li><b>O(n¬≤)</b> ‚Äî quadratic</li>
  </ul>

  <p>We‚Äôll see examples next.</p>

```python
// O(1) - constant time complexity example
let nums = new Set([1,2,3,4]);
console.log(nums.has(3)); // O(1)
console.log(nums.has(5)); // O(1)
```

```python
// O(n) ‚Äî linear time
function sumArray(arr) {
    let sum = 0;
    for (let num of arr) sum += num;
    return sum;
}
console.log(sumArray([1,2,3,4])); // 10

```

```python
// O(n) ‚Äî linear time
function sumArray(arr) {
    let sum = 0;
    for (let num of arr) sum += num;
    return sum;
}
console.log(sumArray([1,2,3,4])); // 10
```

```python
const arr1 = [5, 3, 8, 4, 2];
console.time("O(n log n)");
arr1.sort((a, b) => a - b);
console.timeEnd("O(n log n)");
```

```python
const arr2 = [5, 3, 8, 4, 2];
console.time("O(n¬≤)");
for (let i = 0; i < arr2.length; i++) {
  for (let j = 0; j < arr2.length; j++) {
    // simple comparison
    if (arr2[i] > arr2[j]) {}
  }
}
console.timeEnd("O(n¬≤)");
```

# .sort() ‚Üí O(n log n) because it divides and merges.

# Nested loops ‚Üí O(n¬≤) because it checks every pair.

```python
// üßç O(1) ‚Äî Constant Space
function getFirst(arr) {
  return arr[0]; // Uses fixed memory, no matter array size
}

console.log(getFirst([10, 20, 30]));
// Memory complexity: O(1)
// Explanation: The function uses a constant amount of memory regardless of the input size.
```

```python
// üì¶ O(n) ‚Äî Linear Space
function copyArray(arr) {
  let copy = [...arr]; // Creates a new array of same size
  return copy;
}

console.log(copyArray([1, 2, 3, 4, 5]));
// Memory complexity: O(n)
```

```python
// üßÆ O(n^2) ‚Äî Quadratic Space
function buildMatrix(n) {
  let matrix = [];
  for (let i = 0; i < n; i++) {
    matrix.push(new Array(n).fill(0));
  }
  return matrix;
}

console.log(buildMatrix(3));
// Memory complexity: O(n^2)
```
